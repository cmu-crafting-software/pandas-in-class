```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
all_gdp_data = pd.read_csv('data/gapminder_all.csv', index_col='country')
print('GDP FOR ALL COUNTRIES IN 2007')
print(all_gdp_data.loc[:,'gdpPercap_2007'])

all_gdp_data.boxplot(column='gdpPercap_2007')
```

```python
## life exp

print('LIFE EXPECTANCY FOR ALL COUNTRIES IN 2007')
# print(all_gdp_data.loc[:,'lifeExp_2007'])

continent = all_gdp_data.groupby(by='continent').mean()['lifeExp_2007']

print(continent)

cont_plot = continent.plot(kind='line')
plt.show()
```

```python
lifetimes = all_gdp_data.groupby(by='continent').mean().filter(regex='lifeExp_(198[6-9]|199.|200.)')
print(lifetimes)

lifetimes.plot(kind='line')
plt.show()
```
